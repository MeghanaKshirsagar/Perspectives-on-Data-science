# Handling imbalance dataset with SMOT

```python
# importing required libraries
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
from numpy import where
from imblearn.over_sampling import SMOTE,RandomOverSampler, BorderlineSMOTE, ADASYN
```

```python
# reading the dataset into a dataframe
df = pd.read_csv('winequality-red.csv')
```

```python
# convert the last column quality into categorical values [good, bad]
bins = (2,6.5,8)
labels = ['bad','good']
df['quality'] = pd.cut(df['quality'], bins=bins, labels=labels)
```

```python
#mapping categorical into binary values for easy calculations
df['quality'] = df['quality'].map({'bad':0, 'good':1})
```

```python
#convert the dataframe into feature and predict vectors
# X will hold all the columns except quality
# y will hold the predictable variable 'quality'
X = np.asarray(df.iloc[:,:-1])
y = np.asarray(df['quality'])
```

```python
# caluclate the instances in y vector
counter = Counter(y)
print(counter)
```

```python
# visualise the classes
for label, _ in counter.items():
    row = where(y == label)[0]
    plt.scatter(X[row, 0], X[row, 1], label=str(label))
plt.legend()
plt.show()
```

### Random Oversampling

```python
# create an instance of SMOTE class
oversample = RandomOverSampler()

# oversample the data with fit_resmaple() function
# fit_resample(): takes the feature vector and minorrotu vector and returns the oversampled vector
X, y = oversample.fit_resample(X, y)
```

```python
counter = Counter(y)
print(counter)

# visualise the classes
for label, _ in counter.items():
    row = where(y == label)[0]
    plt.scatter(X[row, 0], X[row, 1], label=str(label))
plt.legend()
plt.show()
```

### SMOTE

```python
# create an instance of SMOTE class
oversample = SMOTE()

# oversample the data with fit_resmaple() function
# fit_resample(): takes the feature vector and minorrotu vector and returns the oversampled vector
X, y = oversample.fit_resample(X, y)
```

```python
counter = Counter(y)
print(counter)
```

```python
# visualise the classes
for label, items in counter.items():
    row = where(y == label)[0]
    plt.scatter(X[row, 0], X[row, 1], label=str(label))
plt.legend()
plt.show()
```
