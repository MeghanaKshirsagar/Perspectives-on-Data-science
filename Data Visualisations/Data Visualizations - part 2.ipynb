## Heatmap

```python
# read the data
df = pd.read_csv('dataset_covid/Covid19_LEACases_Mapped_Historic_Records.csv')
df.head()

df['month'] = pd.DatetimeIndex(df['EventDate']).month
df['year'] = pd.DatetimeIndex(df['EventDate']).year

# extract columns to be plotted
df = df[(df.year == 2021 )]
df = df[(df.month == 4 )]


df['date'] = pd.DatetimeIndex(df['EventDate']).date
date = df['date'].unique()
```

```python
matrix = df.pivot(index='ENGLISH', columns=['EventDate'], values=['P14_100k'])
```

```python
fig, ax = plt.subplots(figsize=(40, 70))
ax = sns.heatmap(matrix,annot=True, cmap='YlGnBu',linewidths=.5, xticklabels=date)
plt.xlabel('Dates in April')
plt.ylabel('Local Electoral Area (LEA)')
plt.title('Incidence rates across LEAs for April 2021 till 19th April, 2021')
plt.savefig('our_plot_name.jpg', format='jpeg', dpi=300)
plt.show()
```
