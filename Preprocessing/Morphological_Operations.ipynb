# Morphological Operations in OpenCv 

```python
# Importing libraries

import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading images

#reading an english digit
img1 = cv2.imread("ImagesMorphology/english.png")

#reading an arabic digit
img2 = cv2.imread("ImagesMorphology/arabic.bmp")

#reverse it to make it uniform like other images
img2 = cv2.bitwise_not(img2)

#reading a japanese digit
img3 = cv2.imread("ImagesMorphology/japanese.png")

#reading a chinese digit
img4 = cv2.imread("ImagesMorphology/chinese.jpg")
```

```python
#Structuring element with numpy and opencv 

# Create a structuring element with numpy
se_numpy = np.ones((3,3), np.uint8)

# Create a structuring element with opencv
se_cv2 = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
```

## Erosion

```python
# erode(img, kernel, iterations): returns an eroded image with the kernel size with specified number of iterations
im1_erosion = cv2.erode(img1, se_cv2, iterations=1)
im2_erosion = cv2.erode(img2, se_cv2, iterations=1)
im3_erosion = cv2.erode(img3, se_cv2, iterations=1)
im4_erosion = cv2.erode(img4, se_cv2, iterations=1)

#display the eroded images
plt.subplot(4,2,1), plt.imshow(img1), plt.title('Original image')
plt.subplot(4,2,2), plt.imshow(im1_erosion), plt.title('Eroded image')
plt.subplot(4,2,3), plt.imshow(img2)
plt.subplot(4,2,4), plt.imshow(im2_erosion)
plt.subplot(4,2,5), plt.imshow(img3)
plt.subplot(4,2,6), plt.imshow(im3_erosion)
plt.subplot(4,2,7), plt.imshow(img4)
plt.subplot(4,2,8), plt.imshow(im4_erosion)
```

## Dilation

```python
# dilate(img, kernel, iterations): returns an eroded image with the kernel size with specified number of iterations
im1_dilate = cv2.dilate(img1, se_cv2, iterations=1)
im2_dilate = cv2.dilate(img2, se_cv2, iterations=1)
im3_dilate = cv2.dilate(img3, se_cv2, iterations=1)
im4_dilate = cv2.dilate(img4, se_cv2, iterations=1)

#display the dilated images
plt.subplot(4,2,1), plt.imshow(img1), plt.title('Original image')
plt.subplot(4,2,2), plt.imshow(im1_dilate), plt.title('Dilated image')
plt.subplot(4,2,3), plt.imshow(img2)
plt.subplot(4,2,4), plt.imshow(im2_dilate)
plt.subplot(4,2,5), plt.imshow(img3)
plt.subplot(4,2,6), plt.imshow(im3_dilate)
plt.subplot(4,2,7), plt.imshow(img4)
plt.subplot(4,2,8), plt.imshow(im4_dilate)
```

## Opening Operation

```python
# morphologyEx(img, operator, kernel): returns resulting image depending upon the operator specified
#MORPH_OPEN: for open operation
#MORPH_CLOSE: for close operation
#MORPH_TOPHAT: for top hat operation
#MORPH_BLACKHAT: for black hat operation
#MORPH_GRADIENT: for morphological gradient

#opening operation
im1_open = cv2.morphologyEx(img1, cv2.MORPH_OPEN, se_cv2)
im2_open = cv2.morphologyEx(img2, cv2.MORPH_OPEN, se_cv2)
im3_open = cv2.morphologyEx(img3, cv2.MORPH_OPEN, se_cv2)
im4_open = cv2.morphologyEx(img4, cv2.MORPH_OPEN, se_cv2)

#display the opened images
plt.subplot(4,2,1), plt.imshow(img1), plt.title('Original image')
plt.subplot(4,2,2), plt.imshow(im1_open), plt.title('Opening image')
plt.subplot(4,2,3), plt.imshow(img2)
plt.subplot(4,2,4), plt.imshow(im2_open)
plt.subplot(4,2,5), plt.imshow(img3)
plt.subplot(4,2,6), plt.imshow(im3_open)
plt.subplot(4,2,7), plt.imshow(img4)
plt.subplot(4,2,8), plt.imshow(im4_open)
```

## Closing Operation

```python
#closing operation
im1_close = cv2.morphologyEx(img1, cv2.MORPH_CLOSE, se_cv2)
im2_close = cv2.morphologyEx(img2, cv2.MORPH_CLOSE, se_cv2)
im3_close = cv2.morphologyEx(img3, cv2.MORPH_CLOSE, se_cv2)
im4_close = cv2.morphologyEx(img4, cv2.MORPH_CLOSE, se_cv2)

#display the resulting images
plt.subplot(4,2,1), plt.imshow(img1), plt.title('Original image')
plt.subplot(4,2,2), plt.imshow(im1_close), plt.title('Closing image')
plt.subplot(4,2,3), plt.imshow(img2)
plt.subplot(4,2,4), plt.imshow(im2_close)
plt.subplot(4,2,5), plt.imshow(img3)
plt.subplot(4,2,6), plt.imshow(im3_close)
plt.subplot(4,2,7), plt.imshow(img4)
plt.subplot(4,2,8), plt.imshow(im4_close)
```

## Morphological Gradient

```python
#gradient operation
im1_gradient = cv2.morphologyEx(img1, cv2.MORPH_GRADIENT, se_cv2)
im2_gradient = cv2.morphologyEx(img2, cv2.MORPH_GRADIENT, se_cv2)
im3_gradient = cv2.morphologyEx(img3, cv2.MORPH_GRADIENT, se_cv2)
im4_gradient = cv2.morphologyEx(img4, cv2.MORPH_GRADIENT, se_cv2)

#display the resulting images
plt.subplot(4,2,1), plt.imshow(img1), plt.title('Original image')
plt.subplot(4,2,2), plt.imshow(im1_gradient), plt.title('Applying Gradient')
plt.subplot(4,2,3), plt.imshow(img2)
plt.subplot(4,2,4), plt.imshow(im2_gradient)
plt.subplot(4,2,5), plt.imshow(img3)
plt.subplot(4,2,6), plt.imshow(im3_gradient)
plt.subplot(4,2,7), plt.imshow(img4)
plt.subplot(4,2,8), plt.imshow(im4_gradient)
```

## Top Hat Transformation

```python
#top hat operation
im1_top = cv2.morphologyEx(img1, cv2.MORPH_TOPHAT, se_cv2)
im2_top = cv2.morphologyEx(img2, cv2.MORPH_TOPHAT, se_cv2)
im3_top = cv2.morphologyEx(img3, cv2.MORPH_TOPHAT, se_cv2)
im4_top = cv2.morphologyEx(img4, cv2.MORPH_TOPHAT, se_cv2)

#display the top hat tranformed images
plt.subplot(4,2,1), plt.imshow(img1), plt.title('Original image')
plt.subplot(4,2,2), plt.imshow(im1_top), plt.title('Top hat transformation')
plt.subplot(4,2,3), plt.imshow(img2)
plt.subplot(4,2,4), plt.imshow(im2_top)
plt.subplot(4,2,5), plt.imshow(img3)
plt.subplot(4,2,6), plt.imshow(im3_top)
plt.subplot(4,2,7), plt.imshow(img4)
plt.subplot(4,2,8), plt.imshow(im4_top)
```

## Black Hat Transformation

```python
#black hat operation
im1_black = cv2.morphologyEx(img1, cv2.MORPH_BLACKHAT, se_cv2)
im2_black = cv2.morphologyEx(img2, cv2.MORPH_BLACKHAT, se_cv2)
im3_black = cv2.morphologyEx(img3, cv2.MORPH_BLACKHAT, se_cv2)
im4_black = cv2.morphologyEx(img4, cv2.MORPH_BLACKHAT, se_cv2)

#display the top hat tranformed images
plt.subplot(4,2,1), plt.imshow(img1), plt.title('Original image')
plt.subplot(4,2,2), plt.imshow(im1_black), plt.title('Black hat transformation')
plt.subplot(4,2,3), plt.imshow(img2)
plt.subplot(4,2,4), plt.imshow(im2_black)
plt.subplot(4,2,5), plt.imshow(img3)
plt.subplot(4,2,6), plt.imshow(im3_black)
plt.subplot(4,2,7), plt.imshow(img4)
plt.subplot(4,2,8), plt.imshow(im4_black)
```
